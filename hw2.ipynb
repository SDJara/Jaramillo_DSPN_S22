```r
#1. Loading data
setwd("~/Documents/GitHub/DataSciencePsychNeuro/Homeworks/lexDat")

data <- read.csv("LexicalData_toclean.csv")

head(data)

#load required packages
require(tidyverse)

#D_RT = reaction time

#2. Data cleaning (save as new column)
#remove commas in D_RT
#delete observations with NA's in D_RT
#make D_RT uniformly postitive 

data2 <- data %>%
  mutate(D_RT_Tidy = D_RT)%>%
  filter(!is.na(D_RT_Tidy))

data2$D_RT_Tidy <- abs(as.numeric(gsub(",","",data2$D_RT_Tidy)))

#check work
head(data2)

head(data2[,], 10)

##Write your response here.
#First action (remove commas): syntatical anomaly, specifically a lexical error
#Second action (delete NAs): coverage anomaly, missing values
#Third action (make all positive): semantic anomaly, integrity constraint violation


#3. Tidy data manipulation
#table with Nobs, mean, and sd of D_RT

table <- data2 %>%
  group_by(Trial)%>%
  summarise(Nobs = n(),
            Mean_RT = mean(D_RT_Tidy),
            SD_RT = sd(D_RT_Tidy))

table

data3 <- data2 %>%
  mutate(Trial_0start = Trial - 1)

head(data3)
head(data3[,], 20)

#4. Plot
plot(data3$Trial_0start,data3$D_RT_Tidy)
#way too much data to get any sense of anything

#randomly sample data points but first we want only data for a certain number of subjects
set.seed(927)

Sub <- sample(data$Sub_ID, size = 15)

data3_grouped <- data3 %>% 
  group_by(Sub_ID)%>%
  filter(Sub_ID == Sub)

plot(data3_grouped$Trial_0start,data3_grouped$D_RT_Tidy)
abline(lm(data3_grouped$D_RT_Tidy ~ data3_grouped$Trial_0start))


#This plot shows reaction time slightly increasing over time (the more trials the longer it takes)
#but overall it's pretty flat and mostly due to noise. Based on this plot, I'd assume no relationship between these factors.







```
